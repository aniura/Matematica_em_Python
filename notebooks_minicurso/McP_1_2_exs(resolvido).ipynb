### Exercício 1:

Modifique o código do penúltimo exemplo para plotar as coordenadas e a trajetória de uma partícula que se move sobre a elipse

$
\begin{cases}
 x(t) & = 4 \cos t\\
 y(t) & = 3 \,\text{sen }t, -\pi \leq t \pi. 
\end{cases}
$

Você deve obter um gráfico similar ao mostrado abaixo.

<img src="imagens/elipse.png" alt="Drawing" style="width: 900px;"/>

```python
#Solução:

#Carregamos o módulo PyPlot
import matplotlib.pyplot as plt

#Carregamos a biblioteca NumPy
import numpy as np

#Se introduzirmos parâmetros nas coordenadas, a trajetória será uma elipse.
#Parâmetros da elipse
a=4
b=3
m = max(a,b)
#Parâmetros da figura
figsize = (14, 8)
colunas = 2
linhas = 1
#Criar os objetos fig e axs
fig, axs = plt.subplots(linhas, colunas, figsize=figsize, layout="constrained")

#Propriedades da figura
fig.suptitle('Coordenadas e trajetória de uma partícula sobre uma elipse')

#Parâmetros do gráfico das coordenadas
axs[0].set_box_aspect(4/3)
axs[0].set_xlim([-np.pi,np.pi])
axs[0].set_ylim(([-m,m]))
axs[0].set_ylabel('x,y')
axs[0].set_xlabel('t')
axs[0].set_title('Coordenadas')

#Parâmetros do gráfico da trajetória
axs[1].set_box_aspect(3/4)
axs[1].set_xlim([-a,a])
axs[1].set_ylim(([-b,b]))
axs[1].set_ylabel('y')
axs[1].set_xlabel('x')
axs[1].set_title(f'Elipse de equação $\\frac{{x^2}}{{{a**2}}} + \\frac{{y^2}}{{{b**2}}} = 1$')

#Plotar as coordenadas
malha_t = np.linspace(-np.pi,np.pi,80)
malha_x = a*np.cos(malha_t)
malha_y = b*np.sin(malha_t)
axs[0].plot(malha_t,malha_x,label='$x(t)$')
axs[0].plot(malha_t,malha_y,label='$y(t)$')

#Plotar a trajetória
axs[1].plot(malha_x,malha_y, label='$(x(t),y(t))$', color = 'tab:green')


#Ativar grid e legendas
axs[0].grid()
axs[0].legend()
axs[1].grid()
axs[1].legend();
```

### Exercício 2: 

Modifique o código do último exemplo para produzir uma figura similar à mostrada abaixo.

**Sugestão**: 

O comando `plt.Rectangle((x, y), a, a, fc='none', ec='tab:orange', lw = 6)` produz um quadrado de 

+ de lado $a$;
+ com lados paralelos aos eixos coordenados;
+ com o canto inferior esquerdo no ponto de coordenadas (x, y);
+ sem cor de recheio;
+ com lados de cor `tab:orange`.

Use esse comando para criar os quadrados como `patches` de tipo `Rectangle` e adicione esses quadrados ao plot usando `add_patch`. 

```python
#Solução 1:

#Carregamos o módulo PyPlot
import matplotlib.pyplot as plt

#Criar os objetos fig e axs
fig, ax = plt.subplots(figsize=(5, 5))

#limite da janela onde vamos desenhar
ax.set_xlim(-4.5, 4.5)
ax.set_ylim(-4.5, 4.5)

#Podemos apagar a caixinha quadrada 
#ax.axis('off')

#Desenhamos os círculos
quadrado1 = plt.Rectangle((-0.3, -0.3), 0.6, 0.6, fc='tab:green', ec='none')

quadrado2 = plt.Rectangle((-2, -2), 4, 4, fc='none', ec='tab:orange', lw = 6)

quadrado3 = plt.Rectangle((-4, -4), 8, 8, fc='none', ec='tab:blue', lw = 6)

#Plotamos os círculos 
ax.add_patch(quadrado1)
ax.add_patch(quadrado2)
ax.add_patch(quadrado3);
```
